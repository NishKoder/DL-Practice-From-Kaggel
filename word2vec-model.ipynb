## Import Libirary <a id="1" />

```python
import pandas as pd
import numpy as np
import os
import gensim
import nltk
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

## Download all nltk required packages <a id="2" />

```python
nltk.download("all")
```

## Read all the files<a id="3"/>

```python
story = []
```

```python
for _,_,filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        story.append(filename)
```

### Experiment with one file<a id="4"/>

```python
f = open(f"/kaggle/input/game-of-thrones-books/{story[0]}", encoding="unicode_escape")
```

```python
corpus=f.read()
```

```python
raw_sentences = sent_tokenize(corpus)
```

### simple preprocess <a id="5"/>

```python
for i in raw_sentences:
    print(simple_preprocess(i))
```

### implement on all files <a id="6" />

```python
story = []
for dirname,_,filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        filepath = os.path.join(dirname, filename)
        with open(filepath, encoding="unicode_escape") as f:
            corpus = f.read()
        raw_sentences = sent_tokenize(corpus)
        for sent in raw_sentences:
            story.append(simple_preprocess(sent))
```

```python
len(story)
```

```python
story[0]
```

## Configure Model<a id="7"/>

```python
model = gensim.models.Word2Vec(window=10, min_count=5, vector_size=150)
```

```python
model.build_vocab(story)
```

```python
model.corpus_count
```

```python
model.epochs
```

```python
model.corpus_total_words
```

## Train Model <a id="8"/>

```python
model.train(
    story, 
    total_examples=model.corpus_count, 
    epochs=model.epochs
)
```
